```python
xdf['total']= (xdf['PHY_TH']+xdf['MTH_TH']+xdf["COM_TH"]+xdf["NEP_TH"]+xdf['CHEM_TH']+xdf['ENG_TH']+\
      xdf['PHY_PR']+xdf['MTH_PR']+xdf["COM_PR"]+xdf["NEP_PR"]+xdf['CHEM_PR']+xdf['ENG_PR']        )  
```

```python
xdf['total']= xdf['total'].astype(object).astype(float)
xdf['percentage']= ((xdf['total']) /f)*100
```

```python
# theory ma fail ly remarks dine ra pass/fail pie chart

xd['Remarks'] = xd.apply(lambda x : 'FAIL' if x['MTH_TH'] <14 or 
                           x['PHY_TH'] < 14  or x['CHEM_TH'] < 14 or x['COM_TH']<14 or x["ENG_TH"]<14 or x["NEP_TH"]<14
                           else 'PASS', axis = 1)

xd['Remarks'].value_counts(dropna = False).plot.pie(colors = ['green', 'red'])
plt.title('overall group results', fontweight = 17, fontsize = 20)
plt.xlabel('Note: absentees counted as failed')
plt.ylabel('P/F Pie Chart')
plt.gcf()
plt.savefig("/home/hari/Desktop/sres/CS12_pi.png")
plt.show()
```
