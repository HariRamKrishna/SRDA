```python
def getgrade(percentage, status):
  if status == 'Fail':
    return 'E'
  if(percentage >= 90):
    return 'O'
  if(percentage >= 80):
    return 'A'
  if(percentage >= 70):
    return 'B'
  if(percentage >= 60):
    return 'C'
  if(percentage >= 40):
    return 'D'
  else :
    return 'E'

xd['grades'] = xd.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1 )

xd['grades'].value_counts()
```
