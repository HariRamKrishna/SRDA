```python
xdf=pd.read_excel(sB11[0])

xdf.columns=['SN','NAME',  'NEP_TH','NEP_PR','ENG_TH','ENG_PR','MTH_TH','MTH_PR','PHY_TH','PHY_PR','CHEM_TH','CHEM_PR','BIO_TH','BIO_PR']
 #'RN','DOB','SYM' omitted

xdxd=xdf

#converting obtained scores to percentage for grading
xdf['PHY_TH']= (100/75)*xdf['PHY_TH']
xdf['MTH_TH']= (100/75)*xdf['MTH_TH']
xdf['BIO_TH']= (100/75)*xdf['BIO_TH']
xdf['NEP_TH']= (100/75)*xdf['NEP_TH']
xdf['CHEM_TH']= (100/75)*xdf['CHEM_TH']
xdf['ENG_TH']= (100/75)*xdf['ENG_TH']

xdf['PHY_PR']= (100/25)*xdf['PHY_PR']
xdf['MTH_PR']= (100/25)*xdf['MTH_PR']
xdf['BIO_PR']= (100/25)*xdf['BIO_PR']
xdf['NEP_PR']= (100/25)*xdf['NEP_PR']
xdf['CHEM_PR']= (100/25)*xdf['CHEM_PR']
xdf['ENG_PR']= (100/25)*xdf['ENG_PR']

del xdf["SN"]



subjects=['PHY_TH','MTH_TH','BIO_TH','NEP_TH','CHEM_TH','ENG_TH','PHY_PR','MTH_PR','BIO_PR','NEP_PR','CHEM_PR','ENG_PR']

#Letter Grading
for z in subjects:
    xdf.loc[ xdf[z] <35, (z+'_GR')] = 'NG'
    xdf.loc[xdf[z] >=90, (z+'_GR')] = 'A+' 
    xdf.loc[ (xdf[z] < 90)  & (xdf[z] >=80), (z+'_GR')] = 'A' 
    xdf.loc[(xdf[z] < 80)  & (xdf[z] >=70), (z+'_GR')] = 'B+' 
    xdf.loc[ (xdf[z] < 70)  & (xdf[z] >=60) , (z+'_GR')] = 'B' 
    xdf.loc[ (xdf[z] < 60)  & (xdf[z] >=50), (z+'_GR')] = 'C+'
    xdf.loc[ (xdf[z] < 50)  & (xdf[z] >=40), (z+'_GR')] = 'C'
    xdf.loc[ (xdf[z] < 40)  & (xdf[z] >=35), (z+'_GR')] = 'D'

#Grade Point
for z in subjects:
    #xdf.loc[ xdf[z] <35, (z+'_GP')] = '***'
    xdf.loc[xdf[z] >=90, (z+'_GP')] = 4 
    xdf.loc[ (xdf[z] < 90)  & (xdf[z] >=80), (z+'_GP')] = 3.6 
    xdf.loc[(xdf[z] < 80)  & (xdf[z] >=70), (z+'_GP')] = 3.2
    xdf.loc[ (xdf[z] < 70)  & (xdf[z] >=60) , (z+'_GP')] = 2.8 
    xdf.loc[ (xdf[z] < 60)  & (xdf[z] >=50), (z+'_GP')] = 2.4
    xdf.loc[ (xdf[z] < 50)  & (xdf[z] >=40), (z+'_GP')] = 2
    xdf.loc[ (xdf[z] < 40)  & (xdf[z] >=35), (z+'_GP')] = 1.6


#

#new NEB final grade system
xdf["CHEM_FINAL"]= (1/4)*xdf["CHEM_PR_GP"]+(3/4)*xdf["CHEM_TH_GP"]
xdf["MTH_FINAL"]= (1/4)*xdf["MTH_PR_GP"]+(3/4)*xdf["MTH_TH_GP"]
xdf["NEP_FINAL"]= (1/4)*xdf["NEP_PR_GP"]+(3/4)*xdf["NEP_TH_GP"]
xdf["ENG_FINAL"]= (1/4)*xdf["ENG_PR_GP"]+(3/4)*xdf["ENG_TH_GP"]
xdf["PHY_FINAL"]=(1/4)*xdf["PHY_PR_GP"]+(3/4)*xdf["PHY_TH_GP"]
xdf["BIO_FINAL"]=(1/4)*xdf["BIO_PR_GP"]+(3/4)*xdf["BIO_TH_GP"]

for v in ['PHY_FINAL','MTH_FINAL','BIO_FINAL','NEP_FINAL','CHEM_FINAL','ENG_FINAL']:
    xdf.loc[ xdf[v] <1.6, (v+'_GR')] = 'NG'
    xdf.loc[xdf[v] >=3.61, (v+'_GR')] = 'A+' 
    xdf.loc[ (xdf[v] <= 3.6)  & (xdf[v] >=3.21), (v+'_GR')] = 'A' 
    xdf.loc[(xdf[v] <=3.2)  & (xdf[v] >=2.81), (v+'_GR')] = 'B+' 
    xdf.loc[ (xdf[v] <=2.8)  & (xdf[v] >=2.41) , (v+'_GR')] = 'B' 
    xdf.loc[ (xdf[v] <=2.4)  & (xdf[v] >=2.01), (v+'_GR')] = 'C+'
    xdf.loc[ (xdf[v] <=2)  & (xdf[v] >=1.61), (v+'_GR')] = 'C'
    xdf.loc[ (xdf[v] ==1.6), (v+'_GR')] = 'D'



xdf["GPA"]= (1/27)*( 2.25*xdf["NEP_TH_GP"]+ 0.75*xdf["NEP_PR_GP"] + 3*xdf["ENG_TH_GP"]+ 1*xdf["ENG_PR_GP"]
             + 3.75*xdf["MTH_TH_GP"]+1.25 *xdf["MTH_PR_GP"]+3.75*xdf["PHY_TH_GP"]+1.25 *xdf["PHY_PR_GP"]+
             3.75*xdf["CHEM_TH_GP"]+1.25 *xdf["CHEM_PR_GP"]+3.75*xdf["BIO_TH_GP"]+1.25*xdf["BIO_PR_GP"])

xd=xdf.sort_values(by="GPA", ascending=0)

xd.reset_index(drop=True, inplace=True)
xd.index += 1

xd

xd.GPA = xd.GPA.round(2)

xd.to_excel('/content/GPARESULT/SCI_BIO_11_GPARESULT.xlsx')
```
